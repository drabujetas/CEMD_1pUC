```python

import numpy as np
import CEMD_1pUC as cemd
import matplotlib.pyplot as plt
import time
```

```python
# Define properties and parameters

a = 300e-9                   # Lattice constant along the "x" axis
b = 300e-9                   # Lattice constant along the "y" axis

n_p = 3.5                    # Particle refractive index divided
n_b = 1                      # Vacuum refractive index
R_p = a/4                    # Particle radius

norm_freq = np.linspace(0.3,0.9,2001)         # Range of normalized frequency = ka/(2pi)
theta = 10*np.pi/180             # Angle of incidence

k = 2*np.pi/a*norm_freq                      # Medium wavevector
```

```python
# Reflectance and transmittance calculation

R_TM = np.zeros((k.shape))
T_TM = np.zeros((k.shape))
R_TE = np.zeros((k.shape))
T_TE = np.zeros((k.shape))
R_CP = np.zeros((k.shape))           # circularly polarized
T_CP = np.zeros((k.shape))

for i in range(k.shape[0]):

    ki = k[i]
    kx = ki*np.sin(theta)
    
    Gb = cemd.GbCalc_1UC(a, b, ki, kx, 0)               # Gb can be loaded
    alp = cemd.get_alpha_Mie(ki/n_b, R_p, n_p, n_b)            # Use the polarizability you wish
    RT = cemd.calc_RT(a, b, Gb, alp, ki, kx, 0)

    R_TM[i] = RT[0]
    T_TM[i] = RT[2]
    R_TE[i] = RT[1]
    T_TE[i] = RT[3]
    
    RT = cemd.calc_RT_pol(a, b, Gb, alp, ki, kx, 0)
    
    R_CP[i] = RT[0]
    T_CP[i] = RT[1]
```

```python
# Plot

plt.figure(1)
plt.plot(norm_freq, R_TM, label = "TM")
plt.plot(norm_freq, R_TE, label = "TE")
plt.xlabel("ka/2pi")
plt.ylabel("R")
plt.legend()
plt.title("Reflectance at 10 deg for linearly polarized light")

plt.figure(2)
plt.plot(norm_freq, R_CP, label = "CP")
plt.xlabel("ka/2pi")
plt.ylabel("R")
plt.legend()
plt.title("Reflectance at 10 deg for circularly polarized")

```
